```python
import sqlite3
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

# Connect to the database
conn = sqlite3.connect('BostonHousing-oddyears.db')

# Execute the SQL query to retrieve the average prices per zip code
query = """
SELECT REPLACE(B.zipcode, ',', '') AS zipcode, AVG(A.BV + A.LV) AS avg_price
FROM Assessments AS A
JOIN Buildings AS B ON A.PID = B.PID
WHERE A.year = 2023
GROUP BY REPLACE(B.zipcode, ',', '')
"""
df_avg_price = pd.read_sql_query(query, conn)

# Convert the 'zipcode' column to object data type and pad with leading zeros
df_avg_price['zipcode'] = df_avg_price['zipcode'].astype(str).str.zfill(5)

# Read the zip code shapefile
shapefile_path = 'ZIP_Codes.shp'
zipcodes = gpd.read_file(shapefile_path)

# Merge the average price data with the zip code shapefile
merged_data = zipcodes.merge(df_avg_price, left_on='ZIP5', right_on='zipcode')

# Plot the shapefile with color coding based on average price
fig, ax = plt.subplots(figsize=(12, 8))
merged_data.plot(column='avg_price', cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True, vmax=1000000)

# Set plot title and axes labels
ax.set_title('Average Price per Zip Code in Boston: 2023')
ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')

# Display the plot
plt.show()
```
